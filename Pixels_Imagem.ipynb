```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carregar a imagem
imagem = cv2.imread('/content/cachorro_bone.jpg')
```

```python
# Convertendo de BGR (formato do OpenCV) para RGB (formato do Matplotlib)
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
# Obter dimensões da imagem
altura, largura, canais = imagem_rgb.shape
```

```python
# Definir a posição do pixel a ser marcada (por exemplo, 100, 50)
pos_y, pos_x = 200, 400
```

```python
# Criar a figura e o eixo
fig, ax = plt.subplots()

# Marcar a posição do pixel
ax.scatter(pos_x, pos_y, color='red', s=100, marker='o', edgecolors='black')  # O ponto marcado

# Exibir a imagem no eixo
ax.imshow(imagem_rgb)
```

```python
# Acessar um pixel específico (linha 200, coluna 400)
pixel = imagem[200, 400]
blue = imagem[200, 400, 0]
green = imagem[200, 400, 1]
red = imagem[200, 400, 2]

print(f"Valor do pixel na posição (200, 400): {pixel}")
print(f"Valor do canal azul: {blue}")
print(f"Valor do canal verde: {green}")
print(f"Valor do canal vermelho: {red}")
```

```python
# Obter dimensões da imagem
altura, largura, canais = imagem.shape
```

```python
# Iterar sobre cada pixel
# Limitando a iteração para a região superior esquerda (200x400) para evitar muitos prints
print("\nIterando sobre a região 200x400 da imagem:")
for y in range(min(200, altura)):
    for x in range(min(400, largura)):
        blue = imagem[y, x, 0]
        green = imagem[y, x, 1]
        red = imagem[y, x, 2]
        print(f"Pixel em ({y}, {x}) - B: {blue}, G: {green}, R: {red}")
```
