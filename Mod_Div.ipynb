# **Recuperando o Resto e o Quociente de uma divisão**

Em uma operação de divisão temos os seguintes termos:
image.png

Além de executar a divisão simples (obtendo um quociente do tipo Real), o python consegue mostrar o Resto e o Quociente de uma divisão como números inteiros.

Recuperar a parte inteira do resto e do quociente é importante para vários tipos de tarefas como por exemplo executar a conversão de um número decimal para binário, calcular intervalos que só podem ser representados por números inteiros, descobrir se um número é par ou ímpar, descobrir números primos, etc.

**Exemplo:**
Imagine que você queira recuperar o resto da divisão e o quociente de 5/2
image.png

```python
# Exemplo: Obtendo o Quociente inteiro de uma divisão
quociente = 5 // 2
quociente
```

```python
# Exemplo: Obtendo o Resto inteiro de uma divisão
resto = 5 % 2
resto
```
