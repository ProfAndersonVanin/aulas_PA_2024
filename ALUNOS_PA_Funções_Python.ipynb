# FUNES

Na programao, funes so blocos de cdigo que realizam determinadas tarefas que normalmente precisam ser executadas diversas vezes dentro de uma aplicao. Quando surge essa necessidade, para que vrias instrues no precisem ser repetidas, elas so agrupadas em uma funo,  qual  dado um nome e que poder ser chamada/executada em diferentes partes do programa.

```python
#Usando Lao de Repetio com retorno
def escalacao():
  listaEscalacao = []
  qtde = int(input("Quantos jogadores vai escalar?"))
  tipo = input("Para que tipo de esportes voc escalar?")
  for i in range(qtde):
    jogador = (input("Escale o Jogador" + str(i+1)))
    listaEscalacao.append(jogador)
  return listaEscalacao, tipo
```

```python
#exemplos de manipulao de retornos usando outras funes que foram realizadas anteriormente, retornando mais de um valor
def calculadora():
  calc = int(input("Calculadora Acionada, qual clculo voc quer fazer? \n Digite 1 para Somar \n Digite 2 para Subtrair \n Digite 3 para Multiplicar \n Digite 4 para Dividir"))
  valor1 = int(input("Digite o Primeiro Valor para fazer o Clculo:"))
  valor2 = int(input("Digite o Segundo Valor para fazer o Clculo:"))
  if calc == 1:
    resul = soma(valor1,valor2)
    tipo = " fez Soma"
  elif calc == 2:
    resul = sub(valor1,valor2)
    tipo =  "fez Subtrao"
  elif calc ==3:
    resul = mult(valor1,valor2)
    tipo = "fez Multiplicao"
  elif calc ==4:
    resul = div(valor1,valor2)
    tipo = " fez Diviso"
  else:
    resul = "invalido"
    tipo = "invalido"
  if resul == "invalido":
    resul = "no digitou um valor vlido pra fazer o clculo"
  return resul, tipo
```

```python
resultado, tipo = calculadora()
```

```python
print("Voc", tipo, "o resultado obtido foi: ", resultado)
```

```python
# Funes com nmero varivel de argumentos
def printVarInfo( arg1, *vartuple ):
   # Imprimindo o valor do primeiro argumento
    print ("O parmetro passado foi: ", arg1)

   # Imprimindo o valor do segundo argumento
    for item in vartuple:
        print ("O parmetro passado foi: ", item)
    return;
```

# **Empacotamento de Parmetros**

```python
def contador(*num):
  print(num)

contador(2,1,7)
contador(8,0)
contador(4,4,7,6,2)
```

**O empacotamento gera uma TUPLA**

```python
def contador(*num):
  print(num)
  tam = len(num)
  print("Foram passados: ",tam," elementos")


contador(2,1,7)
contador(8,0)
contador(4,4,7,6,2)
```

```python
def contador_soma(*valores):
  soma = 0
  for num in valores:
    soma += num
  print("A soma dos valores ",valores,"  igual a ",soma)


contador_soma(2,1,7)
contador_soma(8,0)
contador_soma(4,4,7,6,2)
```

**Exemplo utilizando Listas**

```python
def dobra(lista):
  pos = 0
  while pos<len(lista):
    lista[pos] *= 2
    pos += 1

valores = [7,2,5,0,4]
dobra(valores)
print(valores)
```
